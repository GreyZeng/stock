```python
import akshare as ak
import pandas as pd
import time

def get_single_bond_adjustment_history(bond_code: str):
    """
    è·å–æŒ‡å®šå¯è½¬å€ºçš„è½¬è‚¡ä»·è°ƒæ•´å†å²è®°å½•ã€‚

    Args:
        bond_code (str): éœ€è¦æŸ¥è¯¢çš„å¯è½¬å€ºä»£ç ï¼Œä¾‹å¦‚ "128013"ã€‚
    """
    print(f"ğŸš€ å¼€å§‹æŸ¥è¯¢å¯è½¬å€º {bond_code} çš„è½¬è‚¡ä»·è°ƒæ•´å†å²...")

    # --- æ­¥éª¤ 1: è·å–å€ºåˆ¸ç®€ç§°ï¼Œè®©è¾“å‡ºæ›´å‹å¥½ ---
    bond_name = "æœªçŸ¥ç®€ç§°" # é»˜è®¤å€¼
    try:
        # å€Ÿç”¨å¯è½¬å€ºåˆ—è¡¨æ¥å£å¿«é€ŸæŸ¥è¯¢ç®€ç§°
        bond_list_df = ak.bond_zh_cov()
        name_series = bond_list_df[bond_list_df['å€ºåˆ¸ä»£ç '] == bond_code]['å€ºåˆ¸ç®€ç§°']
        if not name_series.empty:
            bond_name = name_series.iloc[0]
        print(f"  - æŸ¥æ‰¾åˆ°å€ºåˆ¸ç®€ç§°ä¸º: {bond_name}")
    except Exception as e:
        print(f"  âš ï¸ æŸ¥è¯¢å€ºåˆ¸ç®€ç§°æ—¶é‡åˆ°é—®é¢˜: {e}ï¼Œå°†ä½¿ç”¨é»˜è®¤åç§°ã€‚")

    # --- æ­¥éª¤ 2: æŸ¥è¯¢è½¬è‚¡ä»·è°ƒæ•´è®°å½• ---
    try:
        print(f"  - æ­£åœ¨ä»é›†æ€å½•è·å– {bond_code} çš„è°ƒæ•´è®°å½•...")
        adj_log_df = ak.bond_cb_adj_logs_jsl(symbol=bond_code)

        # --- æ­¥éª¤ 3: å¤„ç†ä¸è¾“å‡º ---
        if adj_log_df.empty:
            print(f"\nâœ… æŸ¥è¯¢å®Œæˆï¼šæœªæ‰¾åˆ°å¯è½¬å€º {bond_code} ({bond_name}) çš„ä»»ä½•è½¬è‚¡ä»·è°ƒæ•´è®°å½•ã€‚")
            return

        print("  - æ•°æ®è·å–æˆåŠŸï¼Œæ­£åœ¨æ•´ç†æ ¼å¼...")
        # ä¸ºè¡¨æ ¼æ·»åŠ ä»£ç å’Œç®€ç§°åˆ—ï¼Œä¾¿äºè¯†åˆ«
        adj_log_df['å€ºåˆ¸ä»£ç '] = bond_code
        adj_log_df['å€ºåˆ¸ç®€ç§°'] = bond_name
        
        # è°ƒæ•´åˆ—é¡ºåºï¼Œè®©å…³é”®ä¿¡æ¯æ›´é å‰
        cols_order = [
            'å€ºåˆ¸ä»£ç ', 'å€ºåˆ¸ç®€ç§°', 'è‚¡ä¸œå¤§ä¼šæ—¥', 'ä¸‹ä¿®å‰è½¬è‚¡ä»·', 
            'ä¸‹ä¿®åè½¬è‚¡ä»·', 'æ–°è½¬è‚¡ä»·ç”Ÿæ•ˆæ—¥æœŸ', 'ä¸‹ä¿®åº•ä»·'
        ]
        final_df = adj_log_df[cols_order].sort_values(by='æ–°è½¬è‚¡ä»·ç”Ÿæ•ˆæ—¥æœŸ').reset_index(drop=True)
        
        # åœ¨Jupyterä¸­ç›´æ¥ç¾è§‚åœ°å±•ç¤ºDataFrame
        print("\n--- æŸ¥è¯¢ç»“æœ ---")
        display(final_df) # åœ¨Jupyterä¸­ï¼Œdisplay()æ¯”print()æ›´é€‚åˆå±•ç¤ºDataFrame

        # å¯¼å‡ºåˆ°Excel
        output_filename = f"{bond_code}_{bond_name}_è½¬è‚¡ä»·è°ƒæ•´è®°å½•.xlsx"
        final_df.to_excel(output_filename, index=False)
        print(f"\nğŸ‰ ä»»åŠ¡æˆåŠŸå®Œæˆï¼æ•°æ®å·²ä¿å­˜è‡³æ‚¨çš„å½“å‰ç›®å½•ä¸‹çš„ '{output_filename}' æ–‡ä»¶ä¸­ã€‚")

    except Exception as e:
        print(f"\nâŒ æŸ¥è¯¢æˆ–å¤„ç†è¿‡ç¨‹ä¸­å‘ç”Ÿé”™è¯¯: {e}")
        print("   è¯·æ£€æŸ¥è¾“å…¥çš„å€ºåˆ¸ä»£ç æ˜¯å¦æ­£ç¡®ï¼Œæˆ–ç½‘ç»œè¿æ¥æ˜¯å¦æ­£å¸¸ã€‚")

# --- å¦‚ä½•ä½¿ç”¨ ---
if __name__ == '__main__':
    # è¯·åœ¨è¿™é‡Œè¾“å…¥ä½ æƒ³è¦æŸ¥è¯¢çš„å¯è½¬å€ºä»£ç 
    bond_code_to_query = "128013"  # è¿™æ˜¯ä¸€ä¸ªå¾ˆå¥½çš„ä¾‹å­ï¼Œå› ä¸ºå®ƒæœ‰å¤šæ¡è°ƒæ•´è®°å½•
    
    # è°ƒç”¨å‡½æ•°æ‰§è¡ŒæŸ¥è¯¢
    get_single_bond_adjustment_history(bond_code_to_query)
    
    # ä½ ä¹Ÿå¯ä»¥æŸ¥è¯¢ä¸€ä¸ªæ²¡æœ‰è°ƒæ•´è®°å½•çš„ä¾‹å­
    # print("\n" + "="*50 + "\n")
    # get_single_bond_adjustment_history(bond_code="110088")
```
